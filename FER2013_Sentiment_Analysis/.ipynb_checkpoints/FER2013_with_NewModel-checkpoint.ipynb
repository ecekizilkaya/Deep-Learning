## FER 2013 Sentiment Analysis with New Model

```python
import matplotlib.pyplot as plt
import glob 
import cv2
import os
import random
%matplotlib inline
import warnings 
warnings.filterwarnings('ignore')

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, Input, Dropout, Flatten, Conv2D
from tensorflow.keras.layers import BatchNormalization, Activation, MaxPooling2D
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.utils import plot_model

from IPython.display import SVG, Image
import tensorflow as tf
print("Tensorflow version:", tf.__version__)
```

```python
images = glob.glob("dataset/train/**/**")
for i in range(9):
    image = random.choice(images)
    plt.figure(figsize=(4, 4))
    plt.imshow(cv2.imread(image))
    plt.axis('off')
    plt.show()
```

```python
img_size = 48
batch_size = 64
datagen_train = ImageDataGenerator(horizontal_flip = True)
train_generator = datagen_train.flow_from_directory("dataset/train",
                                               target_size = (img_size, img_size),
                                               color_mode = "grayscale",
                                               batch_size = batch_size,
                                               class_mode = "categorical",
                                               shuffle = True)

datagen_validation = ImageDataGenerator(horizontal_flip = True)
validation_generator = datagen_train.flow_from_directory("dataset/test",
                                               target_size = (img_size, img_size),
                                               color_mode = "grayscale",
                                               batch_size = batch_size,
                                               class_mode = "categorical",
                                               shuffle = True)
```

## Create model.

```python
def Convolution(input_tensor, filters, kernel_size):
    x = Conv2D(filters = filters, kernel_size = kernel_size, padding = "same")(input_tensor)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D(pool_size = (2,2))(x)
    x = Dropout(0.25)(x)

    return x
```

```python
def Dense_f(input_tensor, nodes):
    x = Dense(nodes)(input_tensor)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = Dropout(0.25)(x)

    return x
```

```python
def model_fer(input_shape):
    inputs = Input(input_shape)
    conv_1 = Convolution(inputs, 64,(3,3))
    conv_2 = Convolution(conv_1, 128,(5,5))
    conv_3 = Convolution(conv_2, 512,(3,3))
    conv_4 = Convolution(conv_3, 512,(3,3))

    flatten = Flatten()(conv_4)
    
    dense_1 = Dense_f(flatten,256)
    dense_2 = Dense_f(dense_1,512)

    output = Dense(7,activation = "softmax")(dense_2)
    model = Model(inputs = [inputs], outputs = [output])
    
    opt = Adam(learning_rate=0.0005)

    model.compile(loss = ["categorical_crossentropy"], optimizer = opt, metrics = ["accuracy"])

    return model
```

```python
model = model_fer((48,48,1))
model.summary()
```

## Train the model.

```python
epochs = 15
steps_per_epoch = train_generator.n//train_generator.batch_size
validation_steps = validation_generator.n//validation_generator.batch_size

checkpoint = ModelCheckpoint("new_model.h5", monitor="val_accuracy", save_weights_only=True, save_weights_only=False, mode="max", verbose=1)
reduce_lr = ReduceLROnPlateau(monitor = "val_loss", factor = 0.1, patience = 3, min_lr = 0.00001)
callbacks = [checkpoint,reduce_lr]
```

- If you write "save_weights_only=**True**" in the checkpoint section, you will only save the weights of the model. You will not save the model. If you want to use these saved values again, you will need to create the model and add the weights to this model. 
- If you want to save the entire model, you should use the "save_weights_only=**False**" option. In this way, you can easily load the model as <br> 
 **model = load_model("model_complete.h5")**.

```python
history = model.fit(
x = train_generator,
steps_per_epoch = steps_per_epoch,
epochs = epochs,
validation_data = validation_generator,
validation_steps = validation_steps,
callbacks  = callbacks)
```

## If you have created and saved the model, you can use that model again.

```python
# Used to load a saved model.
from tensorflow.keras.models import load_model
model = load_model("new_model.h5")
```

## Ploting accuracy and loss.

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Model Accuracy")
plt.legend(["Train","Validation"],loc = "upper left")
plt.subplots_adjust(top=1.0,bottom=0.0,right=0.95,hspace=0.25,wspace=0.35)
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Model Loss")
plt.legend(["Train","Validation"],loc = "upper left")
plt.subplots_adjust(top=1.0,bottom=0.0,right=0.95,hspace=0.25,wspace=0.35)
```

## Let's make prediction.

```python
from PIL import Image
import numpy as np
from tensorflow.keras.applications.vgg16 import preprocess_input

def process_image(image_path):
    img = Image.open(image_path).resize((48, 48)).convert("L")  # Grayscale
    img = np.array(img)  
    img = img / 255.0  # Normalizing (0-1)
    img = np.expand_dims(img, axis=-1)  # (48, 48, 1)
    img = np.expand_dims(img, axis=0)  # (1, 48, 48, 1)
    return img
```

```python
image_path = "PrivateTest_9488237.jpg"  
img = process_image(image_path)
```

```python
preds = model.predict(img)
preds
```

```python
image_classes = ["angry", "disgust", "fear", "happy", "neutral", "sad", "surprise"]
result = np.argmax(preds[0])
print(image_classes[result])
```
